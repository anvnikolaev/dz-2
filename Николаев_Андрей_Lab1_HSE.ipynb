## 1. Парсеки

Напишите функцию, которая принимает на вход:
- np.array - набор расстояний от Земли до звёзд (в световых годах)
- k - количество значений, которое нам нужно

Выведите топ-k самых удаленных звёзд от нас в порядке убывания разрастания, расстояния выведите в парсеках (1 парсек приблизительно равен 3.262 световым годам)

```python
def find_top_k_longest_distance_stars(stars, k):
  sorted_stars = np.sort(stars)[::-1]
  sorted_stars = sorted_stars / 3.262
  return sorted_stars[:k:]
```

```python
# Просто пример, можно придумать свой
print(find_top_k_longest_distance_stars(np.array([4.37, 8.6, 4.24, 4.37, 5.96, 7.8, 8.6, 7.9]), 4))
```

## 2. Оцениваем размер галактики

Напишите функцию, которая принимает на вход:
- np.array - набор масс звёзд в галактике
- k - количество максимальных по массе звёзд

Функция должна вывести два числа - первое - сумму масс первых k звёзд по массе, второе - сумму масс остальных звёзд.

```python
def count_k_masses(stars, k):
  sorted_stars = np.sort(stars)[::-1]
  sum_heaviest = np.sum(sorted_stars[:k])
  sum_other = np.sum(sorted_stars) - sum_heaviest
  return sum_heaviest, sum_other
```

```python
# Просто пример, можно придумать свой
print(count_k_masses(np.array([150, 200, 300, 250, 400, 100, 200]), 2))
```

```python
import warnings
import numpy as np
from scipy.stats import zscore
```

```python
def NaNs(column):
    median = column.median()
    new_column = column.fillna(median)
    return new_column
# Заменяет NaN на медиану столбца

def outliers(column):
    z_scores = zscore(column)
    median = column.median()
    outliers = np.abs(z_scores) > 3
    column.loc[outliers] = median
    return column
# Заменяем выбросы на медиану столбца

warnings.simplefilter(action='ignore', category=FutureWarning)
# Код выводил очень много предупреждений, поэтому я их отключил

df = merged

df = df.loc[:, df.isnull().sum() <= 3]
# Уберем все столбцы, в которых количество пропусков больше, чем 3

df = df.apply(NaNs)
df = df.apply(outliers)
# Все выбросы и пропуски заменяем на медиану столбца.

pd.set_option('display.float_format', '{:.2f}'.format)
display(df['Доходы бюджета'], df['Расходы бюджета'])
# Вручную убедились, что в столбцах с общими доходами и расходами ничего не поменялось

display(df)
df_1 = df
```
